```python
# Taken from: https://stackoverflow.com/questions/51070985/find-out-the-percentage-of-missing-values-in-each-column-in-the-given-dataset
# This function will return a table with the number of missing values and the percentage of missing values for each column in the dataframe
# The table will be sorted by the percentage of missing values in descending order
def missing_zero_values_table(df):
        zero_val = (df == 0.00).astype(int).sum(axis=0)
        mis_val = df.isnull().sum()
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        mz_table = pd.concat([zero_val, mis_val, mis_val_percent], axis=1)
        mz_table = mz_table.rename(
        columns = {0 : 'Zero Values', 1 : 'Missing Values', 2 : '% of Total Values'})
        mz_table['Total Zero Missing Values'] = mz_table['Zero Values'] + mz_table['Missing Values']
        mz_table['% Total Zero Missing Values'] = 100 * mz_table['Total Zero Missing Values'] / len(df)
        mz_table['Data Type'] = df.dtypes
        mz_table = mz_table[
            mz_table.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns and " + str(df.shape[0]) + " Rows.\n"      
            "There are " + str(mz_table.shape[0]) +
              " columns that have missing values.")
        return mz_table

missing_zero_values_table(df)
```

```python
# non_mercado_pago_payment_methods
import re
import unicodedata

def clean_column_name(name):
    name = name.lower().strip()  # Convertir a minúsculas y quitar espacios extra
    name = unicodedata.normalize("NFKD", name).encode("ASCII", "ignore").decode("utf-8")  # Eliminar tildes
    name = re.sub(r"\s+", "_", name)  # Reemplazar espacios con guion bajo
    return name

def extract_payment_methods(payment_list):
    if not isinstance(payment_list, list):
        return {}
    return {clean_column_name(p["description"]): True for p in payment_list}

payment_df = df["non_mercado_pago_payment_methods"].apply(extract_payment_methods).apply(pd.Series, dtype='bool').fillna(False)
df = pd.concat([df, payment_df], axis=1)
```

For the 'garantia' (warranty) variable, we can observe that there's a description indicating whether the product has a warranty or not. A function has been created to attempt to normalize this variable and determine in which cases the warranty applies or doesn't apply. Regular expressions have been used to determine applicability. In some cases, it states '6 months,' '1 year,' 'yes,' 'does not apply,' 'by default,' etc. (More sophisticated methods could be employed). In some cases, it's unknown (for example: NUESTRA REPUTACION), and these have been marked as indeterminado

```python
import re
import unicodedata

# Función para normalizar el texto (eliminar tildes y convertir a minúsculas)
def normalizar_texto(texto):
    if isinstance(texto, str):
        texto = texto.lower()  # Convertir a minúsculas
        texto = ''.join(c for c in unicodedata.normalize('NFD', texto) if unicodedata.category(c) != 'Mn')  # Eliminar tildes
        texto = re.sub(r'[^\w\s]', '', texto)  # Eliminar caracteres especiales
    return texto

# Expresiones regulares para detectar periodos de tiempo en garantía
patron_tiempo = re.compile(r'(\d+)\s*(dias|día|mes|meses|año|años|semanas|semana|ano)', re.IGNORECASE)

# Actualizar la lista de palabras clave para la garantía
keywords_aplica = ["garantia", "cubre", "respaldado", "siempre que", "proteccion", "defectos de fabricacion", "aplica", "si", "garantizado", "por fallas", "fabricacion", "defecto"]
keywords_no_aplica = ["no","desconocido", "no aplica", "sin garantia", "no se aceptan devoluciones", "sin", "no cubre", "no garantia", "no respaldado", "no proteccion", "no defectos de fabricacion", "no aplica", "no garantizado", "no por fallas", "no fabricacion", "no defecto"]

# Función mejorada para clasificar la garantía
def clasificar_garantia_mejorado(texto):
    if isinstance(texto, str):
        texto_norm = normalizar_texto(texto)  # Normalizar el texto
        
        # Si menciona directamente "sí", se clasifica como Aplica
        if "si" in texto_norm:
            return "aplica"
        
        # Si menciona una cantidad de tiempo (ejemplo: "6 meses", "1 año"), se clasifica como Aplica
        if patron_tiempo.search(texto_norm):
            return "aplica"
        
        # Verificar palabras clave de garantía aplicable
        if any(kw in texto_norm for kw in keywords_aplica):
            return "aplica"
        
        # Verificar palabras clave de garantía no aplicable
        if any(kw in texto_norm for kw in keywords_no_aplica):
            return "no_aplica"
    
    return "indeterminado"

# Aplicar la nueva clasificación al dataset
df["garantia_aplica"] = df["warranty"].apply(clasificar_garantia_mejorado)
```

```python
import pandas as pd

def convert_to_category(df, category_cols):
    """
    Convierte columnas en tipo 'category'.
    """
    for col in category_cols:
        if col in df.columns:
            df[col] = df[col].astype("category")

    return df

def convert_to_numeric(df, numeric_cols):
    """
    Convierte columnas en tipo 'float' (o 'int' si es posible).
    """
    for col in numeric_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')  # Convierte y pone NaN en errores

    return df

# Definir los tipos de columnas
category_cols = [
    'condition', 'warranty', 'buying_mode', 'currency_id', 'seller_country', 
    'seller_state', 'seller_city', 'shipping_mode', 'parent_item_id', 
    'category_id', 'seller_id', 'official_store_id', 'video_id', 'status', 'garantia_aplica','listing_type_id'
]

numeric_cols = [
    'initial_quantity', 'available_quantity', 'sold_quantity', 
    'original_price', 'base_price', 'price'
]

# Aplicar funciones al DataFrame 
data = convert_to_category(data, category_cols)
data = convert_to_numeric(data, numeric_cols)
```

```python
from collections import defaultdict

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import spearmanr
from scipy.cluster import hierarchy
from scipy.spatial.distance import squareform

from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.inspection import permutation_importance
from sklearn.model_selection import train_test_split

X = data.drop(columns=['condition'])
y = data['condition']

feature_selc_model = RandomForestClassifier(n_estimators=100, random_state=42)
feature_selc_model.fit(X, y)
```

```python
result = permutation_importance(feature_selc_model, X, y, n_repeats=5, random_state=42, n_jobs=-1)
perm_sorted_idx = result.importances_mean.argsort()
```

```python
import numpy as np
import matplotlib.pyplot as plt

tree_importance_sorted_idx = np.argsort(feature_selc_model.feature_importances_)
tree_indices = np.arange(0, len(feature_selc_model.feature_importances_)) + 0.5

fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.barh(tree_indices, feature_selc_model.feature_importances_[tree_importance_sorted_idx], height=0.7, color='blue', alpha=0.7, edgecolor='black')
ax1.set_yticks(tree_indices)
ax1.set_yticklabels(X.columns[tree_importance_sorted_idx])
ax1.set_ylim((0, len(feature_selc_model.feature_importances_)))

plt.show()
```
